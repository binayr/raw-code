### 115. Version to number

```python
def version_to_num(ver):
    x = [int(i) for i in ver.split('.')]
    exp = 0
    out = 0
    for i in x[::-1]:
        out += 10**exp * i
        exp += 1
    return out
```

### 117. Binary tree right view

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        out = {}
        def right(root, level):
            if root is not None:
                if level not in out:
                    out[level] = root.val
                if root.right is not None:
                    right(root.right, level+1)
                if root.left is not None:
                    right(root.left, level+1)
        right(root, 0)
        return out.values()

s = Solution()
```

### 118. Implement Trie

```python
class TrieNode:
    def __init__(self, char = ""):
        self.char = char
        self.children = {}
        self.is_end = False
        # self.counter = 0


class Trie:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        """
        Inserts a word into the trie.
        """
        node = self.root
        for char in word:
            if char in node.children:
                node = node.children[char]
            else:
                new_node = TrieNode(char)
                node.children[char] = new_node
                node = new_node
        node.is_end = True
        # node.counter += 1
        

    def search(self, word: str) -> bool:
        """
        Returns if the word is in the trie.
        """
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        
        # Reached at the end of word
        # return True if word is present, i.e is_end = True else False
        return node.is_end
        

    def startsWith(self, prefix: str) -> bool:
        """
        Returns if there is any word in the trie that starts with the given prefix.
        """
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True
        
```

### 119. Minimum Size Subarray Sum

```python
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        out = float('inf')
        temp = 0
        i,j = 0,0
        n = len(nums)

        while i < n:
            #print('...',i, j, temp)
            if temp >= target:
                #print(i,j,temp, out)
                out = min(j-i, out)
                temp -= nums[i]
                i += 1
                #print(temp)
            elif j >= n:
                temp -= nums[i]
                i += 1
            else:
                temp += nums[j]
                j += 1
            if out == 1: break
        return 0 if out == float('inf') else out

s = Solution()
```

### 120. Design Add and Search Words Data Structure

```python
class WordDictionary(object):

    def __init__(self):
        self.set = set()
        self.sizeDict = defaultdict(set)
        self.letterDict = defaultdict(lambda: defaultdict(set))

    def addWord(self, word):
        """
        :type word: str
        :rtype: None
        """
        self.sizeDict[len(word)].add(word)
        self.set.add(word)
        for i,char in enumerate(word):
            self.letterDict[i][char].add(word)

    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """
        if word in self.set:
            return True
        charIdx = [i for i,char in enumerate(word) if char != "."]
        contCheck = self.sizeDict[len(word)]
        for i in charIdx:
            contCheck = contCheck & self.letterDict[i][word[i]]
        return bool(len(contCheck))
```

### 122. Basic Calculator II

```python
class Solution:
    def calculate(self, s: str) -> int:
        s = s.strip()
        stack = []
        num = 0
        operator = '+'
        for i in range(len(s)):
            char = s[i]
            if char.isdigit():
                num = num * 10 + int(char)
            if i == len(s) - 1 or char == '+' or char == '-' \
                or char == '*' or char == '/':
                if operator == '+':
                    stack.append(num)
                if operator == '-':
                    stack.append(-num)
                if operator == '*':
                    stack.append(stack.pop() * num)
                if operator == '/':
                    stack.append(int(stack.pop() / num))
                # update for the next operation
                operator = char
                num = 0
        res = 0
        while len(stack) > 0:
            res += stack.pop()
        return res

s = Solution()
                
```

### 123. Majority element II

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> List[int]:
        counter = {}
        out = set()
        n = len(nums)
        threshold = n//3
        
        for i in nums:
            if i in out: continue
            if i in counter:
                counter[i] += 1
            else:
                counter[i] = 1
            
            if counter[i] > threshold: out.add(i)
        return out
```

### 124. Kth Smallest Element in a BST

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def kthSmallest(self, root, k):
        """
        :type root: TreeNode
        :type k: int
        :rtype: int
        """
        def inorder(r):
            return inorder(r.left) + [r.val] + inorder(r.right) if r else []
    
        return inorder(root)[k - 1]
```

### 125. lowest common ancestor of a binary tree

```python
class Solution:

    def __init__(self):
        # Variable to store LCA node.
        self.ans = None

    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        def recurse_tree(current_node):

            # If reached the end of a branch, return False.
            if not current_node:
                return False

            # Left Recursion
            left = recurse_tree(current_node.left)

            # Right Recursion
            right = recurse_tree(current_node.right)

            # If the current node is one of p or q
            mid = current_node == p or current_node == q

            # If any two of the three flags left, right or mid become True.
            if mid + left + right >= 2:
                self.ans = current_node

            # Return True if either of the three bool values is True.
            return mid or left or right

        # Traverse the tree
        recurse_tree(root)
        return self.ans
```

### 126. product of array except self

### 127. Search a 2D Matrix II

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        def binarySearch(nums, target):
            start, end = 0, len(nums) -1
            while start < end:
                mid = (start+end-1)//2
                if nums[mid] > target:
                    end = mid
                elif nums[mid] < target:
                    start = mid+1
                elif nums[mid] == target:
                    return True
            else:
                return False
        for row in matrix:
            if row[0] < target < row[-1] and binarySearch(row, target): return True
        else:
            return False

s = Solution()
```

### 128. Single Number

```python
class Solution:
    def singleNumber(self, nums: List[int]) -> List[int]:
        mapper = {}
        for i in nums:
            if i in mapper:
                mapper[i] += 1
            else:
                mapper[i] = 1
        return list(filter(lambda x: mapper[x]==1, mapper))

s = Solution()
```

### 130. H Index II

```python
class Solution:
    def hIndex(self, citations: List[int]) -> int:
        n = len(citations)
        
        start = 0 
        end = n 
        
        ans = min(citations[0], n)
        
        while end - start > 1: 
            m = (end + start)//2
            if citations[m] > n - m: 
                ans = max(ans, n - m)
                end = m
            else: 
                ans = max(ans, citations[m])
                start = m
            
        return ans
```

### 131. Peeking Iterator

```python
# Below is the interface for Iterator, which is already defined for you.
#
# class Iterator:
#     def __init__(self, nums):
#         """
#         Initializes an iterator object to the beginning of a list.
#         :type nums: List[int]
#         """
#
#     def hasNext(self):
#         """
#         Returns true if the iteration has more elements.
#         :rtype: bool
#         """
#
#     def next(self):
#         """
#         Returns the next element in the iteration.
#         :rtype: int
#         """

class PeekingIterator:
    def __init__(self, iterator):
        """
        Initialize your data structure here.
        :type iterator: Iterator
        """
        self.it = iterator
        self.peeked = None
        

    def peek(self):
        """
        Returns the next element in the iteration without advancing the iterator.
        :rtype: int
        """
        if not self.peeked:
            if not self.it.hasNext():
                return None
            self.peeked = self.it.next()
        return self.peeked
        

    def next(self):
        """
        :rtype: int
        """
        if self.peeked:
            peeked, self.peeked = self.peeked, None # Assign self.peeked to peeked and reset self.peeked
            return peeked
        return self.it.next()
        

    def hasNext(self):
        """
        :rtype: bool
        """
        return self.peeked is not None or self.it.hasNext()
        

# Your PeekingIterator object will be instantiated and called as such:
# iter = PeekingIterator(Iterator(nums))
# while iter.hasNext():
#     val = iter.peek()   # Get the next element but not advance the iterator.
#     iter.next()         # Should return the same value as [val].
```

### 134. Maximum Product of Word Lengths

```python
class Solution:
    def maxProduct(self, words: List[str]) -> int:
        mapper = [(set(i), len(i)) for i in words]
        out = 0
        i = 0
        while i < len(mapper):
            for j in range(i+1, len(mapper)):
                if not mapper[i][0].intersection(mapper[j][0]):
                    out = max(mapper[i][1] * mapper[j][1], out)
            i += 1

        return out

s = Solution()
```
